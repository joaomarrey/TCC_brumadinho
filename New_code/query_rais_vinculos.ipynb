```python
#query = """
#         SELECT ano, sigla_uf, id_municipio, id_municipio_trabalho,
#                    AVG(valor_remuneracao_media), AVG(valor_remuneracao_janeiro), AVG(valor_remuneracao_fevereiro), AVG(valor_remuneracao_marco),
#                    AVG(valor_remuneracao_abril), AVG(valor_remuneracao_maio), AVG(valor_remuneracao_junho), AVG(valor_remuneracao_julho), 
#                    AVG(valor_remuneracao_agosto), AVG(valor_remuneracao_setembro), AVG(valor_remuneracao_outubro), AVG(valor_remuneracao_novembro), 
#                    AVG(valor_remuneracao_dezembro), 
#         
#         FROM `basedosdados.br_me_rais.microdados_vinculos`
#         
#         WHERE sigla_uf = 'MG' AND ano >= 2002
#         
#         GROUP BY ano, id_municipio
#         
#         ORDER BY ano, id_municipio
#         
#         """

query = """
         SELECT ano AS ANO, 
                sigla_uf AS SIGLA_UF,  
                id_municipio AS COD_MUN,
                AVG(valor_remuneracao_dezembro) AS REMUNERACAO_DEZEMBRO, 
         
         FROM `basedosdados.br_me_rais.microdados_vinculos`
         
         WHERE sigla_uf = 'MG' AND ano >= 2002 AND valor_remuneracao_dezembro IS NOT NULL AND valor_remuneracao_dezembro <> 0
         
         GROUP BY ano, sigla_uf, id_municipio
         
         ORDER BY ano, id_municipio
         
         """
```

```python
rais_vinculos_dez_df = bd.read_sql(query=query,
                    billing_project_id=project_name)
```

```python
rais_vinculos_dez_df.head()
```

```python
query = """
         SELECT ano AS ANO, 
                sigla_uf AS SIGLA_UF,  
                id_municipio AS COD_MUN,
                AVG(valor_remuneracao_media) AS REMUNERACAO_MEDIA,
         
         FROM `basedosdados.br_me_rais.microdados_vinculos`
         
         WHERE sigla_uf = 'MG' AND ano >= 2002 AND valor_remuneracao_media IS NOT NULL AND valor_remuneracao_media <> 0
         
         GROUP BY ano, sigla_uf, id_municipio
         
         ORDER BY ano, id_municipio
         
         """
```

```python
rais_vinculos_med_df = bd.read_sql(query=query,
                    billing_project_id=project_name)
```

```python
rais_vinculos_med_df.head()
```

```python
rais_vinculos_df = pd.merge(rais_vinculos_dez_df, rais_vinculos_med_df, how='left', on=['ANO', 'SIGLA_UF', 'COD_MUN'])
```

```python
rais_vinculos_df.to_csv(os.path.join(manipulated_data_path, 'rais_vinculos_municipios.csv'), index=False)
```
