```python
import basedosdados as bd
import pandas as pd
from paths import manipulated_data_path, project_name
import os
```

```python
query = """
          SELECT ano AS ANO, 
                sigla_uf AS SIGLA_UF,  
                id_municipio AS COD_MUN,
                SUM(quantidade_vinculos_ativos) AS VINCULOS_ATIVOS,
                SUM(quantidade_vinculos_clt) AS VINCULOS_CLT,
                SUM(quantidade_vinculos_estatutarios) AS VINCULOS_ESTATUTARIOS,
                COUNT(*) AS NUM_FIRMAS, 

         FROM `basedosdados.br_me_rais.microdados_estabelecimentos`

         WHERE sigla_uf = 'MG' AND ano >= 2002
         
         GROUP BY ano, sigla_uf, id_municipio
         
         ORDER BY ano, id_municipio
         """
```

```python
rais_estabelecimentos_df = bd.read_sql(query=query,
                    billing_project_id=project_name)
```
