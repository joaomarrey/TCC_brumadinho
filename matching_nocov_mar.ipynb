```python
import pandas as pd
```

```python
df = pd.read_csv("./Data/dados_municipais_sal_emprego_rais_1985_2022.csv")
df = df[(df["ano"]>=2011) & (df["ano"]<=2022)]
df
```

```python
pop = pd.read_csv("./Data/codigos_pop_municipios.csv")
pop = pop.rename({" CODIGO RAIS ": "id_municipio", ' POPULAÇÃO ESTIMADA ': "pop", "NOME DO MUNICÍPIO": "nome"}, axis=1)
pop
```

```python
df = pd.merge(df, pop.loc[:, ["id_municipio", "pop", "nome"]], how="left", on=["id_municipio"])
df
```

```python
df.dropna(inplace=True)
df
```

```python
# matching mariana
treated = [3140001, 3105707, 3157401, 3155009
          ]

df_mar = df[df["id_municipio"].isin([3140001, 3145901, 3131703, 3144805, 3155702, 3125101 # Mariana
                                    #   , 3105707, 3164472, 3133907, 3120839, 3102506, 3141504 # Barra Longa
                                    #   , 3157401, 3112653, 3121001, 3106804, 3123304, 3140407 # santa cruz do escalvado
                                    #   , 3155009, 3164902, 3165602, 3147501, 3124609, 3156205 # rio doce
                                   ])]
df_mar
```

```python
df_mar["treated_group"] = 0
df_mar.loc[df["id_municipio"].isin(treated), "treated_group"] = 1
```

```python
df_mar["treated_group"]
```

```python
df_mar["treated_after_treatment"] = 0
df_mar.loc[(df_mar["treated_group"]) & (df_mar["ano"] >= 2015), "treated_after_treatment"] = 1
```

```python
df_mar.loc[(df_mar["treated_group"]) & (df_mar["ano"] >= 2019), "treated_after_treatment"] = 1
["treated_after_treatment"]
```
