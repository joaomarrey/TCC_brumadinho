```python
import pandas as pd
```

```python
for i in range(2015, 2022):
    df = pd.read_excel(f"./Data/pop_{i}.xls")
    df["ano"] = i
    if i == 2015:
        pop = df
    else:
        pop = pd.concat([pop, df])

pop
```

```python
pop.dropna(inplace=True)
```

```python
pop["id_municipio"] = pop["COD. MUNIC"] + (pop["COD. UF"] * 100000)
```

```python
pop = pop.rename({'POPULAÇÃO ESTIMADA': "pop", "NOME DO MUNICÍPIO": "nome"}, axis=1)
```

```python
pop.to_csv("./Data/pop_final.csv", index=False)
```
