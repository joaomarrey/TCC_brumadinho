```python
import pandas as pd
```

```python
df = pd.read_csv("./Data/dados_municipais_sal_emprego_rais_1985_2022.csv")
df = df[(df["ano"]>=2015) & (df["ano"]<=2022)]
df
```

```python
pop = pd.read_csv("./Data/codigos_pop_municipios.csv")
pop = pop.rename({" CODIGO RAIS ": "id_municipio", ' POPULAÇÃO ESTIMADA ': "pop", "NOME DO MUNICÍPIO": "nome"}, axis=1)
pop
```

```python
df = pd.merge(df, pop.loc[:, ["id_municipio", "pop", "nome"]], how="left", on=["id_municipio"])
df
```

```python
df.dropna(inplace=True)
df
```
